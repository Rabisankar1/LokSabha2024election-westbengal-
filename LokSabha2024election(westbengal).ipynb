```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("lok_sabha_2024_sample_cleaned.csv")
print(df.head())
```

```python
df.info()
```

Party-wise Total Votes

# Party wise Total Votes

```python
party_votes = df.groupby("party")["votes"].sum().sort_values(ascending = False)
plt.figure()
party_votes.plot(kind = 'bar')
plt.title("Total Votes by Party")
plt.xlabel("Party")
plt.ylabel("Votes")
plt.xticks(rotation = 45)
plt.show()
```

# Winners vs Losers Count

```python
sns.countplot(x = 'winner', data = df)
plt.title("winners vs Others")
plt.show()
```

# Margin Analysis

```python
plt.figure()
sns.histplot(df["margin"], kde = True)
plt.title("Winning Mergin Distribution")
plt.xlabel("Margin")
plt.show()
```

# Gender Participation

```python
sns.countplot(x = "gender", data = df)
plt.title("Gender Distribution of Canditates")
plt.show()
```

# Top Canditates by Vote Share 

```python
top_canditates = df.sort_values("vote_share_%", ascending = False).head(10)
sns.barplot(x = "vote_share_%", y = "candidate", data=top_canditates)
plt.title("Top 10 Canditates by Vote Share")
plt.show()
```

# How does vote share vary acrosss  parties?

```python
plt.figure()
sns.boxplot(x = "party", y="vote_share_%", data=df)
plt.xticks(rotation=45)
plt.title("How does vote share vary acrosss parties?")
plt.show()
```

# Does age affect vote share?

```python
plt.figure()
sns.scatterplot(x="age", y="vote_share_%", data=df)
plt.title("Does age affect vote share")
plt.show()
```

# How is vote share distributed among candidates?

```python
plt.figure()
sns.histplot(df["vote_share_%"], kde=True)
plt.title("How is vote share distributed among candidates?")
plt.xlabel("vote share (%)")
plt.show()
```

# Parties win the most seats

```python
party_wins = df[df["winner"] == True]["party"].value_counts()
plt.figure()
party_wins.plot(kind = 'bar')
plt.title("Which parties  win the most seates?")
plt.xlabel("")
```
